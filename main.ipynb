```python
from diffusers.models.unets.unet_2d import UNet2DModel


class LargeConvDenoiserNetwork(nn.Module):
    def __init__(
        self,
        in_channels: int = 3,
        out_channels: int = 3,
        channels: list[int] = [64, 128, 256, 512, 1024],
        layers_per_block: int = 2,
        downblock: str = 'ResnetDownsampleBlock2D',
        upblock: str = 'ResnetUpsampleBlock2D',
        add_attention: bool = True,
        attention_head_dim: int = 64,
        low_condition: bool = False,
        timestep_condition: bool = True,
        global_skip_connection: bool = True,
        num_class_embeds: int | None = None,
    ):
        super().__init__()
        self.low_condition = low_condition
        self.timestep_condition = timestep_condition
        self.global_skip_connection = global_skip_connection
        self.divide_factor = 2 ** len(channels)

        in_channels = 2 * in_channels if self.low_condition else in_channels

        self.backbone = UNet2DModel(
            in_channels=in_channels,
            out_channels=out_channels,
            block_out_channels=channels,
            layers_per_block=layers_per_block,
            down_block_types=tuple(downblock for _ in range(len(channels))),
            up_block_types=tuple(upblock for _ in range(len(channels))),
            add_attention=add_attention,
            attention_head_dim=attention_head_dim,
            num_class_embeds=num_class_embeds,
        )

    def padding(self, x):
        _, _, W, H = x.shape
        desired_width = (
            (W + self.divide_factor - 1) // self.divide_factor
        ) * self.divide_factor
        desired_height = (
            (H + self.divide_factor - 1) // self.divide_factor
        ) * self.divide_factor

        # Calculate the padding needed
        padding_w = desired_width - W
        padding_h = desired_height - H

        return F.pad(x, (0, padding_h, 0, padding_w), mode="constant", value=0), W, H

    def remove_padding(self, x, W, H):
        return x[:, :, :W, :H]

    def forward(self, x_t, t):

        # add padding to fit nearest value divisible by self.divide_factor
        x_in, W, H = self.padding(x_t)

        model_output = self.backbone(
            x_in,
            timestep=t if self.timestep_condition else 0,
        ).sample

        model_output = self.remove_padding(model_output, W, H)

        if self.global_skip_connection:
            model_output[:, :3] = model_output[:, :3] + x_t

        return model_output  # pred_x_0
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchvision.utils import save_image
from tqdm import tqdm
import os
import tempfile

from pytorch_fid import fid_score
from torchmetrics.image.inception import InceptionScore


def generate_images(model, device, num_images=5000, batch_size=100):
    try:
        model.model.to(device)
        model.model.eval()
    except AttributeError:
        model.to(device)
        model.eval() 
    generated_images = []
    
    with torch.no_grad():
        for _ in tqdm(range(num_images // batch_size), desc="Generating images"):
            labels = torch.randint(0, 10, (batch_size, )).to(device)
            samples = model.sample(use_ema=True, labels=labels).to(device)  # Your model should implement this
            samples = (samples.clamp(-1, 1) + 1) / 2  # Rescale to [0,1]
            generated_images.append(samples.cpu())
    
    return torch.cat(generated_images, dim=0)

def compute_fid(model, dataset, device, num_images=1000, batch_size=100):
    try:
        model.model.to(device)
        model.model.eval()
    except AttributeError:
        model.to(device)
        model.eval() 
    
    # Generate images and save to temp dir
    with tempfile.TemporaryDirectory() as gen_dir, tempfile.TemporaryDirectory() as real_dir:
        gen_imgs = generate_images(model, device, num_images=num_images, batch_size=batch_size)
        
        # Save generated images
        for i, img in enumerate(gen_imgs):
            save_image(img, os.path.join(gen_dir, f"{i}.png"))
        
        # Save real images
        real_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
        count = 0
        for imgs, _ in real_loader:
            for img in imgs:
                save_image(img, os.path.join(real_dir, f"{count}.png"))
                count += 1
                if count >= num_images:
                    break
        
        fid_value = fid_score.calculate_fid_given_paths([gen_dir, real_dir], batch_size=batch_size, device=device, dims=2048)
        
    return fid_value

# === Function to evaluate Inception Score ===
def compute_inception_score(model, device, num_images=1000, batch_size=100):
    try:
        model.model.to(device)
        model.model.eval()
    except AttributeError:
        model.to(device)
        model.eval() 

    images = generate_images(model, device, num_images=num_images, batch_size=batch_size)
    inception_score = InceptionScore().to(device)

    for i in range(0, num_images, batch_size):
        batch = images[i:i+batch_size].to(device)
        batch = (batch.clamp(0, 1) * 255).to(torch.uint8)
        inception_score.update(batch)

    mean, std = inception_score.compute()
    return mean.item(), std.item()


def compute_metrics(model, dataset, device, num_images=1000, batch_size=100):
    try:
        model.model.to(device)
        model.model.eval()
    except AttributeError:
        model.to(device)
        model.eval() 

    # Generate images and save to temp dir
    with tempfile.TemporaryDirectory() as gen_dir, tempfile.TemporaryDirectory() as real_dir:
        gen_imgs = generate_images(model, device, num_images=num_images, batch_size=batch_size)
        inception_score = InceptionScore().to(device)

        for i in range(0, num_images, batch_size):
            batch = gen_imgs[i:i+batch_size].to(device)
            batch = (batch.clamp(0, 1) * 255).to(torch.uint8)  # Fix here
            inception_score.update(batch)

        mean, std = inception_score.compute()
        
        # Save generated images
        for i, img in enumerate(gen_imgs):
            save_image(img, os.path.join(gen_dir, f"{i}.png"))
        
        # Save real images
        real_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
        count = 0
        for imgs, _ in real_loader:
            for img in imgs:
                save_image(img, os.path.join(real_dir, f"{count}.png"))
                count += 1
                if count >= num_images:
                    break
        
        fid_value = fid_score.calculate_fid_given_paths([gen_dir, real_dir], batch_size=batch_size, device=device, dims=2048)
        
    return fid_value, mean.item(), std.item() 
```
